```python
import os
#import kaggle #conda install -c kaggle 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
with open('oscar_rotten.pkl', 'rb') as f:
  df_oscar_rotten=pickle.load(f)
```

```python
df_oscar_rotten.info()
```

```python
df_nominee = df_oscar_rotten.loc[df_oscar_rotten['nominee']==True]
```

```python
content_rating = df_nominee.content_rating .value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 39 - Classificação indicativa dos indicados ao Oscar")
sns.barplot(x = content_rating.index, y = content_rating);
```

```python
production_company = df_nominee.groupby('production_company')
production_company_nominee = production_company.nominee.sum().sort_values(ascending=False).head(30)
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation=75)
plt.title("Gráfico 40 - Ranking de produtoras por filmes indicados ao Oscar")
sns.barplot(x=production_company_nominee.index, y=production_company_nominee);
```

```python
tomatometer_status = df_nominee.tomatometer_status.value_counts()
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 41 - Status dos indicados no Rotten Tomatoes")
sns.barplot(x=tomatometer_status.index, y=tomatometer_status);
```

```python
graph=sns.displot(df_nominee['tomatometer_rating'], kde=True, bins=30, height=8, aspect=2)
graph.fig.subplots_adjust(top=.9)
graph.fig.suptitle('Gráfico 42 - Distribução das críticas positivas no Rotten Tomatoes dos indicados do Oscar');
```

```python
tomatometer_count=df_nominee.sort_values(by=['tomatometer_count'],ascending=False).head(30) 
    
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title('Gráfico 43 - Ranking de filmes pela contagem de críticas no Rotten Tomatoes')
sns.barplot(x=tomatometer_count.movie_title, y=tomatometer_count.tomatometer_count);
```

```python
audience_status = df_nominee.audience_status.value_counts()
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 44 - Status da audiência no Rotten Tomatoes dos filmes indicados ao Oscar")
sns.barplot(x=audience_status.index, y=audience_status);
```

```python
graph=sns.displot(df_nominee['audience_rating'], kde=True, bins=30, height=8, aspect=2)
graph.fig.subplots_adjust(top=.9)
graph.fig.suptitle('Gráfico 45 -  Distribuição das avaliações da audiência no Rotten Tomatoes dos filmes indicados ao Oscar');
```

```python
tomatometer_top_critics_count=df_nominee.sort_values(by=['tomatometer_top_critics_count'],ascending=False).head(30) 
 
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 46 - Ranking dos filmes indicados ao Oscar pela crítica dos autores principais")
sns.barplot(x=tomatometer_top_critics_count.movie_title, y=tomatometer_top_critics_count.tomatometer_top_critics_count);
```
