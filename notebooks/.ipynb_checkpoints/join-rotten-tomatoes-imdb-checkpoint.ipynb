```python
import os
import pandas as pd
#from pandas_profiling import ProfileReport
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
df_imdb = pd.read_csv('df_imdb_tratado.csv', sep=',', na_values=[''], keep_default_na=False, low_memory=False) # dtype={'startYear':int}, na_values='\\N',
```

```python
df_imdb.head()
```

```python
df_imdb.startYear.unique()
```

```python
df_imdb.dropna(subset=['startYear'], inplace=True)
```

```python
df_imdb.startYear.unique()
```

```python
df_imdb['startYear']=df_imdb['startYear'].astype(int)
```

```python
df_imdb.head()
```

```python
df_rotten = pd.read_csv('df_rotten_tomatoes_tratado.csv',sep=',')
```

```python
df_rotten.head()
```

```python
df_imdb.head()
```

```python
df_rotten.head()
```

```python
df_imdb_rotten = df_imdb.merge(df_rotten, left_on=['primaryTitle','startYear'],right_on=['movie_title','year'], how='outer')
```
