```python
import os
import kaggle #conda install -c kaggle 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle

%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
with open('rotten.pkl', 'rb') as f:
    df_rotten_tomatoes=pickle.load(f)
```

```python
df_rotten_tomatoes.info()
```

```python
df_rotten_tomatoes.head()
```

```python
content_rating = df_rotten_tomatoes.content_rating.value_counts()
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 12 - Distribuição da classificação indicativa")
sns.barplot(x=content_rating.index, y=content_rating);
```

```python
tomatometer_status = df_rotten_tomatoes.tomatometer_status.value_counts()
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 13 - Status das produções")
sns.barplot(x=tomatometer_status.index, y=tomatometer_status);
```

```python
graph14=sns.displot(df_rotten_tomatoes['tomatometer_rating'], kde=True, bins=30, height=8, aspect=2)
graph14.fig.subplots_adjust(top=.9)
graph14.fig.suptitle('Gráfico 14 - Distribuição do percentual de críticas positivas')
graph14;
```

```python
content_rating = df_rotten_tomatoes.groupby('content_rating')
tomatometer_count_cr = content_rating.tomatometer_count.sum()
tomatometer_count_cr

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 15 - Distribuição dos percentual de críticas positivas por classificação indicativa")
sns.barplot(x=tomatometer_count_cr.index, y=tomatometer_count_cr);
```

```python
tomatometer_status = df_rotten_tomatoes.groupby('tomatometer_status')
tomatometer_count_ts = tomatometer_status.tomatometer_count.sum()
tomatometer_count_ts

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 16 - Distribuição das críticas por status da produção")
sns.barplot(x=tomatometer_count_ts.index, y=tomatometer_count_ts);
```

```python
audience_status = df_rotten_tomatoes.audience_status.value_counts()
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 17 - Distribuição dos status das produções indicados pela audiência")
sns.barplot(x=audience_status.index, y=audience_status);
```

```python
graph18=sns.displot(df_rotten_tomatoes['audience_rating'], kde=True, bins=30, height=8, aspect=2)
graph18.fig.subplots_adjust(top=.9)
graph18.fig.suptitle('Gráfico 18 - Distribuição das avaliações da audiência')
graph18;
```

```python
content_rating = df_rotten_tomatoes.groupby('content_rating')
audience_count_cr = content_rating.audience_count.sum()
audience_count_cr 

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 19 - Distribuição dos votos do público por classificação indicativa")
sns.barplot(x=audience_count_cr.index, y=audience_count_cr);
```

```python
audience_status = df_rotten_tomatoes.groupby('audience_status')
audience_count_ts = audience_status.audience_count.sum()
audience_count_ts

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 20 - Distribuição de críticas do público por status indicado pela audiência")
sns.barplot(x=audience_count_ts.index, y=audience_count_ts);
```

```python
content_rating = df_rotten_tomatoes.groupby('content_rating')
tomatometer_top_critics_count_cr = content_rating.tomatometer_top_critics_count.sum()
tomatometer_top_critics_count_cr 

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 21 - Distribuição das críticas dos críticos principais por classificação indicativa")
sns.barplot(x=tomatometer_top_critics_count_cr.index, y=tomatometer_top_critics_count_cr);
```

```python
tomatometer_status = df_rotten_tomatoes.groupby('tomatometer_status')
tomatometer_top_critics_count_ts = tomatometer_status.tomatometer_top_critics_count.sum()
tomatometer_top_critics_count_ts

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 22 - Distribuição das críticas dos críticos principais por status da produção")
sns.barplot(x=tomatometer_top_critics_count_ts.index, y=tomatometer_top_critics_count_ts);
```

```python
graph23 = sns.displot(df_rotten_tomatoes['tomatometer_top_critics_count'], kde=True, bins=30, height=8, aspect=2);
graph23.fig.subplots_adjust(top=.9)
graph23.fig.suptitle('Gráfico 23 - Distribuição contagem de críticas de críticos principais')
graph23;
```

```python
production_company = df_rotten_tomatoes.groupby('production_company')
pop_production_company = production_company.audience_count.sum().sort_values(ascending=False).head(30) 
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation=75)
plt.title("Gráfico 24 - Produtoras com mais críticas da audiência")
sns.barplot(x=pop_production_company.index, y=pop_production_company);
```

```python
rotten = df_rotten_tomatoes.loc[df_rotten_tomatoes['tomatometer_status'].isin(['Rotten'])]
rotten_production_company = rotten.groupby('production_company')
graf_rotten_production_company = rotten_production_company.tomatometer_status.count().sort_values(ascending=False).head(30) 
```

```python
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 25 - Produtoras com mais filmes com status Rotten")
sns.barplot(x=graf_rotten_production_company.index, y=graf_rotten_production_company);
```

```python
fresh = df_rotten_tomatoes.loc[df_rotten_tomatoes['tomatometer_status'].isin(['Fresh'])]
fresh_production_company = fresh.groupby('production_company')
graf_fresh_production_company = fresh_production_company.tomatometer_status.count().sort_values(ascending=False).head(30) 
```

```python
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 26 - Produtoras com mais filmes com status Fresh")
sns.barplot(x=graf_fresh_production_company.index, y=graf_fresh_production_company);
```
