```python
import os
import pandas as pd
#from pandas_profiling import ProfileReport
import seaborn as sns
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
df_imdb = pd.read_csv('df_imdb_tratado.csv', sep=',', na_values='\\N', low_memory=False)
```

```python
df_oscar=pd.read_csv('df_oscar_tratado.csv')
```

```python
df_rotten = pd.read_csv('df_rotten_tomatoes_tratado.csv',sep=',')
```

```python
df_imdb_oscar = df_imdb.merge(df_oscar, left_on=['primaryTitle','startYear'], right_on=['film','year_film'], how='outer')
```

```python
df_imdb_oscar_rotten = df_imdb_oscar.merge(df_rotten, left_on=['primaryTitle','startYear'],right_on=['movie_title','year'], how='outer')
```
