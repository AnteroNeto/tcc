```python
import os
#import kaggle #conda install -c kaggle 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle

%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
with open('oscar_imdb.pkl', 'rb') as f:
    df_imdb_oscar=pickle.load(f)
```

```python
df_imdb_oscar.info()
```

```python
df_nominee = df_imdb_oscar.loc[df_imdb_oscar['nominee']==True]
```

```python
titleType = df_nominee.titleType.value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 31 - Filmes indicados ao Oscar por tipo da produção")
sns.barplot(x=titleType.index, y=titleType);
```

```python
genres = df_nominee.genres.value_counts().sort_values(ascending=False).head(30) 
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 32 - Fimes indicados ao Oscar por gênero")
sns.barplot(x=genres.index, y=genres);
```

```python
graph = sns.displot(df_nominee['averageRating'], kde=True, bins=30, height=8, aspect=2)
graph.fig.subplots_adjust(top=.9)
graph.fig.suptitle('Gráfico 33 - Distribuição das médias das avaliações dos filmes indicados');
```

```python
numVotes=df_nominee.sort_values(by=['numVotes'],ascending=False).head(30)
```

```python
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 34 - Número de votos por indicado ao Oscar")
sns.barplot(x=numVotes.primaryTitle, y=numVotes.numVotes);
```

```python
df_winner = df_imdb_oscar.loc[df_imdb_oscar['winner']==True]
```

```python
titleType_winner = df_winner.titleType.value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Gráfico 35 - Vencedores do Oscar por tipo de produção")
sns.barplot(x=titleType_winner .index, y=titleType_winner);
```

```python
genres = df_winner.genres.value_counts().sort_values(ascending=False).head(30) 
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 36 - Filmes vencedores do Oscar por gênero")
sns.barplot(x=genres.index, y=genres);
```

```python
graph = sns.displot(df_winner['averageRating'], kde=True, bins=30, height=8, aspect=2)
graph.fig.subplots_adjust(top=.9)
graph.fig.suptitle('Gráfico 37 - Distribuição das médias das avaliações dos filmes vencedores');
```

```python
numVotes=df_winner.sort_values(by=['numVotes'],ascending=False).head(30)
```

```python
plt.figure(figsize=(16,6))
plt.xticks(rotation=75)
plt.title("Gráfico 38 - Número de votos por vencedor ao Oscar")
sns.barplot(x=numVotes.primaryTitle, y=numVotes.numVotes);
```
