```python
import os
import kaggle #conda install -c kaggle 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle

%matplotlib inline
```

```python
os.chdir("../tratados")
os.getcwd()
```

```python
with open('oscar.pkl', 'rb') as f:
    df_oscars=pickle.load(f)
```

```python
df_oscars.info()
```

```python
oscars = df_oscars.groupby('film')
oscars_nominee =oscars.nominee.sum().sort_values(ascending=False).head(30)
plt.figure(figsize=(20,10))
plt.xticks(rotation=75)
plt.title("Gráfico 29 - Filmes mais com indicações ao Oscar")
sns.barplot(x=oscars_nominee.index, y=oscars_nominee);
```

```python
oscars_winner =oscars.winner.sum().sort_values(ascending=False).head(30)
plt.figure(figsize=(20,10))
plt.xticks(rotation=75)
plt.title("Gráfico 30 - Ranking de maiores vencedores do Oscar")
sns.barplot(x=oscars_winner.index, y=oscars_winner);
```
